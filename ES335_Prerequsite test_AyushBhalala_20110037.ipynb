### Ans. 1

#### To perform a matrix multiplication between an (n, k) and a (k, m) matrix, you need n * k * m multiplications and n * k * (m - 1) additions. 

#### The resulting product matrix will have dimensions (n, m), with n rows and m columns. To calculate each element of the product matrix, you perform a dot product between a row of the first matrix and a column of the second matrix. Each dot product involves k multiplications and k-1 additions. Hence, the total number of operations: 
#### Multiplications: n * m * k 
#### Additions: n * m * (k - 1)

### Ans 2

```python
import time

# Define the matrices
matrix_a = [[1, 2, 3], [4, 5, 6]]
matrix_b = [[7, 8, 9], [10, 11, 12], [13, 14, 15]]

# Function for list of lists multiplication
def list_multiplication(matrix_a, matrix_b):
    result = [[0 for _ in range(len(matrix_b[0]))] for _ in range(len(matrix_a))]
    for i in range(len(matrix_a)):
        for j in range(len(matrix_b[0])):
            for k in range(len(matrix_b)):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]
    return result

# Time the list of lists multiplication
start_time_list = time.time()
result_list = list_multiplication(matrix_a, matrix_b)
end_time_list = time.time()
time_list = end_time_list - start_time_list
```

```python
import numpy as np

# Function for NumPy multiplication
def numpy_multiplication(matrix_a, matrix_b):
    return np.dot(matrix_a, matrix_b)

# Time the NumPy multiplication
start_time_numpy = time.time()
result_numpy = numpy_multiplication(matrix_a, matrix_b)
end_time_numpy = time.time()
time_numpy = end_time_numpy - start_time_numpy
```

```python
# Print results and timing
print("List of lists multiplication:")
print(result_list)
print(f"Time taken: {time_list:.5f} seconds")

print("NumPy multiplication:")
print(result_numpy)
print(f"Time taken: {time_numpy:.5f} seconds")

# Comparison
if time_list > time_numpy:
    print("NumPy is faster!")
else:
    print("List of lists is surprisingly faster (for small matrices)!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the domain
x = np.arange(0.5, 100.1, 0.5)

# Define the functions
def f1(x):
    return x

def f2(x):
    return x**2

def f3(x):
    return (x**3) / 100

def f4(x):
    return np.sin(x)

def f5(x):
    return np.sin(x) / x

def f6(x):
    return np.log(x)

def f7(x):
    return np.exp(x)

# Create the subplots
fig, axes = plt.subplots(3, 2, figsize=(12, 8))

# Plot the functions
axes[0, 0].plot(x, f1(x), label='y = x')
axes[0, 1].plot(x, f2(x), label='y = x^2')
axes[1, 0].plot(x, f3(x), label='y = (x^3)/100')
axes[1, 1].plot(x, f4(x), label='y = sin(x)')
axes[2, 0].plot(x, f5(x), label='y = sin(x)/x')
axes[2, 1].plot(x, f6(x), label='y = log(x)')

# Add labels and title
axes[0, 0].set_ylabel('y')
axes[0, 1].set_ylabel('y')
axes[1, 0].set_ylabel('y')
axes[1, 1].set_ylabel('y')
axes[2, 0].set_ylabel('y')
axes[2, 1].set_ylabel('y')
axes[0, 0].set_title('y = x')
axes[0, 1].set_title('y = x^2')
axes[1, 0].set_title('y = (x^3)/100')
axes[1, 1].set_title('y = sin(x)')
axes[2, 0].set_title('y = sin(x)/x')
axes[2, 1].set_title('y = log(x)')

# Add legend
axes[0, 0].legend()
axes[0, 1].legend()
axes[1, 0].legend()
axes[1, 1].legend()
axes[2, 0].legend()
axes[2, 1].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd

# Generate random matrix
matrix = np.random.uniform(1, 2, size=(20, 5))

# Convert to Pandas DataFrame
df = pd.DataFrame(matrix, columns=["a", "b", "c", "d", "e"])

# Add new column with sum of existing columns
df["f"] = df[["a", "b", "c", "d", "e"]].sum(axis=1)

# Add new column based on condition
df["g"] = np.where(df["f"] < 8, "LT8", "GT8")

# Count rows with "LT8" in column "g"
lt8_count = df["g"].value_counts()["LT8"]

# Calculate standard deviation of "f" for "LT8" and "GT8" groups
lt8_std = df[df["g"] == "LT8"]["f"].std()
gt8_std = df[df["g"] == "GT8"]["f"].std()

# Print results
print("Number of rows with 'LT8' in 'g':", lt8_count)
print("Standard deviation of 'f' for 'LT8':", lt8_std)
print("Standard deviation of 'f' for 'GT8':", gt8_std)
```
